## Leyendo un json

```python
import json
import pandas as pd 
import requests
```

```python
res = requests.get('https://raw.githubusercontent.com/juanxo90/basico_python/main/data/raw_nyc_phil.json')
```

```python
with open(f"raw_nyc_phil.json", 'wb') as archivo:
    for i in res:
        archivo.write(i)
```

```python
with open('/content/raw_nyc_phil.json') as f:
    d = json.load(f)
```

```python
dphil = pd.json_normalize(d['programs'])
dphil.head(3)
```

```python
works = pd.json_normalize(data=d['programs'], record_path= 'works', 
                  meta = ["id", "orchestra", "programID", "season"])
works.head(3)
```

```python
solistas = pd.json_normalize(data=d['programs'], record_path=['works', 'soloists'],
                             meta = ['id', 'orchestra'])
solistas.head(3)
```

```python
data = pd.merge(works[['workTitle', 'conductorName','composerName','id', 'ID']],
         solistas, on=['id'])
```

```python
data.head(3)
```
